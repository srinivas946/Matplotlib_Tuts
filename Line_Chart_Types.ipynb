### <center>Line Charts</center>

```python
import matplotlib.pyplot as plt # import matplotlib.pyplot package as short name plt
# use this line to display the charts on jupyter notebook
%matplotlib inline  
```

**Basic Line Plot**

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8'] # Create x axis list
y = [10, 12, 15, 3, 5] # Create y axis as list
plt.plot(x, y)    # Use plot method to draw the graph and give input x and y values
plt.show()        # Data Visualization
```

**change the plot color and plot style**

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8'] # Create x axis list
y = [10, 12, 15, 3, 5] # Create y axis as list
plt.plot(x, y, color='g', linestyle='dashdot') # Use color & linestyle attributes
plt.show()        # Data Visualization
```

**Label x - axis and y- axis & Add title**

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8'] # Create x axis list
y = [10, 12, 15, 3, 5] # Create y axis as list
plt.plot(x, y, color='r', linestyle='dashed') # Use color & linestyle attributes
plt.xlabel('IPAddress') # use xlabel to name x-axis
plt.ylabel('Total Hits') # use y label to name y-axis
plt.title('Data Visualization Related to IPAddress') # use title to set the title
plt.show()        # Data Visualization
```

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8'] # Create x axis list
y = [10, 12, 15, 3, 5] # Create y axis as list
plt.plot(x, y, color='c', linestyle='dashed') # Use color & linestyle attributes
plt.xlabel('IPAddress') # use xlabel to name x-axis
plt.ylabel('Total Hits') # use y label to name y-axis
plt.title('Data Visualization Related to IPAddress') # use title to set the title
plt.show()        # Data Visualization
```

**Add text inside the plot**

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8'] 
y = [10, 12, 15, 3, 5] 
plt.plot(x, y, color='c', linestyle='dashed') 
plt.xlabel('IPAddress') 
plt.ylabel('Total Hits') 
plt.title('Data Visualization Related to IPAddress') 

# create a dictionary with font properties
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
# use text method to add text to plot
# x --> x-axis position
# y --> y-axis position
# s --> String (text) to be displayed
# fontdict --> Accepts a dictionary with font properties
plt.text(x=1.5, y=5, s='Matplotlib', fontdict=font)
plt.show()        
```

***for more text properties refer the link***
***https://matplotlib.org/api/text_api.html#matplotlib.text.Text***

**Using alpha**<br/>
Using alpha we can increase or decrease the thickness of image or text

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8'] 
y = [10, 12, 15, 3, 5] 
plt.plot(x, y, color='y', linestyle='solid', alpha=0.5) # Use alpha range between 0 to 1
plt.xlabel('IPAddress') 
plt.ylabel('Total Hits') 
plt.title('Data Visualization Related to IPAddress')
plt.show()
```

**Draw Multiple plot**

```python
# plot one data
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8'] 
y = [10, 12, 15, 3, 5]

# plot two data (x-axis remains same)
q = [2, 14, 8, 5, 9]
plt.plot(x, y, color='b', linestyle='dashdot') # first plot
plt.plot(x, q, color='r', linestyle='dashed')  # second plot
plt.xlabel('IPAddress') 
plt.ylabel('Total Hits') 
plt.title('Data Visualization Related to IPAddress')
plt.show()
```

**Add Legend for plot**

```python
# plot one data
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8'] 
y = [10, 12, 15, 3, 5]

# plot two data (x-axis remains same)
q = [2, 14, 8, 5, 9]

# Add label to the plot and store it in another variable
line1, = plt.plot(x, y, color='b', linestyle='dashdot', label='plot1') # first plot

# Add label to the plot and store it in another variable
line2, = plt.plot(x, q, color='r', linestyle='dashed', label='plot2')  # second plot
plt.xlabel('IPAddress') 
plt.ylabel('Total Hits') 
plt.title('Data Visualization Related to IPAddress')

# Add legend
# title --> Add the title of the Legend has to be displayed
# handles --> number of plots
# bbox_to_anchor --> position of the legend
# shadow --> If requires shadow use True otherwise make it False

plt.legend(title='IPAddress', handles=[line1, line2], bbox_to_anchor=(0.75, 0, 0.5, 1), shadow=True)
plt.show()
```

**How to Save Image**

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8'] # Create x axis list
y = [10, 12, 15, 3, 5] # Create y axis as list
plt.plot(x, y)    # Use plot method to draw the graph and give input x and y values
# use .savefig() method to save the figure
# fname --> filename of the image along with the extension(.jpg, .jpeg, .png, .pdf etc...)
# dpi --> dots per inches
# bbox_inches --> set the image properly while saving
plt.savefig(fname='filename_to_save.png', dpi=500, bbox_inches='tight')
print('Image Created')
```

**Displaying values in line plot**

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8'] 
y = [10, 12, 3, 5, 15] 
plt.plot(x, y, color='c', linestyle='dashed') 
plt.xlabel('IPAddress') 
plt.ylabel('Total Hits') 
plt.title('Data Visualization Related to IPAddress')
for i, j in enumerate(y):
    plt.text(x=i, y=j+0.3, s=str(j))
plt.show()
```

**xlim and ylim**

```python
# In the above figure we can observe that Displayed value are merging with the border of graph to
# avoid that we can use xlim (x limit value), ylim (y limit value)
```

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8'] 
y = [10, 12, 3, 5, 15] 
plt.plot(x, y, color='c', linestyle='dashed') 
plt.xlabel('IPAddress') 
plt.ylabel('Total Hits') 
plt.ylim(0, 20) # set y limit to 20
plt.title('Data Visualization Related to IPAddress')
for i, j in enumerate(y):
    plt.text(x=i, y=j+0.3, s=str(j))
plt.show()
```

```python
# But if you want to make ylim value dynamic refer below script
```

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8'] 
y = [10, 12, 3, 5, 15] 
plt.plot(x, y, color='c', linestyle='dashed') 
plt.xlabel('IPAddress') 
plt.ylabel('Total Hits') 

# get the maximum value of list using max() function
# add extra 5 or 10 count to the result of max() function
plt.ylim(0, max(y)+5)
plt.title('Data Visualization Related to IPAddress')
for i, j in enumerate(y):
    plt.text(x=i, y=j+0.3, s=str(j))
plt.show()
```
