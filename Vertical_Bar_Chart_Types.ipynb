```python
import matplotlib.pyplot as plt # import matplotlib.pyplot package as short name plt
# use this line to display the charts on jupyter notebook
%matplotlib inline
```

<b>Basic Bar Chart</b>

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8'] # Create x axis list
y = [10, 12, 15, 3, 5] # Create y axis as list
plt.bar(x, y)    # Use plot method to draw the graph and give input x and y values
plt.show()        # Data Visualization
```

<b>Change Bar Color</b>

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8']
y = [10, 12, 15, 3, 5]
# Use color parameter to change bar color eg: color='r' for red color
plt.bar(x, y, color='r', alpha=0.6)    
plt.show()
```

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8']
y = [10, 12, 15, 3, 5]
# pass list of colors to color parameter to get multi colors for bar plot
plt.bar(x, y, color=['blue', 'red', 'green', 'yellow', 'orange'], alpha=0.5)    
plt.show()
```

<b>Change Bar Width</b>

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8']
y = [10, 12, 15, 3, 5]
# Use width parametre to adjust the width of bar
plt.bar(x, y, color=['skyblue', 'red', 'green', 'yellow', 'orange'], alpha=0.5, width=0.5)    
plt.show()
```

<b>Add X Lables and Y Labels & Title</b>

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8']
y = [10, 12, 15, 3, 5]
plt.bar(x, y, color=['skyblue', 'orange', 'green', 'yellow', 'pink'], alpha=0.5, width=0.5)
plt.xlabel('IPAddress') # add label for x-axis
plt.ylabel('Count') # add label for y-axis
plt.title('IP Address Statistics') # add title to the bar graph
plt.show()
```

<b>Displaying Values in Bar Plot</b>

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8']
y = [10, 12, 15, 3, 5]
plt.bar(x, y, color=['skyblue', 'orange', 'green', 'yellow', 'pink'], alpha=0.5, width=0.5)
plt.xlabel('IPAddress')
plt.ylabel('Count')
plt.title('IP Address Statistics')
# add text on the bar plot
for i, j in enumerate(y):
    plt.text(x=i, y=j+0.3, s=str(j))
plt.show()
```

<b>Removing Spines from right and top</b>

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8']
y = [10, 12, 15, 3, 5]
plt.bar(x, y, color=['skyblue', 'orange', 'green', 'yellow', 'pink'], alpha=0.5, width=0.5)
plt.xlabel('IPAddress')
plt.ylabel('Count')
plt.title('IP Address Statistics', pad=25)
plt.gca().spines['right'].set_visible(False) # remove right spine
plt.gca().spines['top'].set_visible(False) # remove left spine
for i, j in enumerate(y):
    plt.text(x=i, y=j+0.3, s=str(j))
plt.show()
```
