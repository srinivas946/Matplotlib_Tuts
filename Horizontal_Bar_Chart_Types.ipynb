```python
import matplotlib.pyplot as plt # import matplotlib.pyplot package as short name plt
# use this line to display the charts on jupyter notebook
%matplotlib inline
```

<b>Basic Horizontal Bar Chart</b>

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8'] # Create x axis list
y = [10, 12, 15, 3, 5] # Create y axis as list
plt.barh(x, y)    # Use plot method to draw the graph and give input x and y values
plt.show()        # Data Visualization
```

<b>Change Bar Color</b>

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8']
y = [10, 12, 15, 3, 5]
# Use color parameter to change bar color eg: color='r' for red color
plt.barh(x, y, color='green', alpha=0.6)    
plt.show()
```

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8']
y = [10, 12, 15, 3, 5]
# Use color parameter to change bar color eg: color='r' for red color
plt.barh(x, y, color=['blue', 'pink', 'red', 'green','black'], alpha=0.6)    
plt.show()
```

<b>Add X Lables and Y Labels & Title</b>

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8']
y = [10, 12, 15, 3, 5]
plt.barh(x, y, color=['skyblue', 'orange', 'green', 'yellow', 'pink'], alpha=0.5)
plt.xlabel('Count') # add label for x-axis
plt.ylabel('IPAddress') # add label for y-axis
plt.title('IP Address Statistics') # add title to the bar graph
plt.show()
```

<b>Displaying Values in Bar Plot</b>

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8']
y = [10, 12, 15, 3, 5]
plt.barh(x, y, color=['skyblue', 'orange', 'green', 'yellow', 'pink'], alpha=0.5)
plt.xlabel('Count')
plt.ylabel('IPAddress')
plt.title('IP Address Statistics')
# add text on the bar plot
for i, j in enumerate(y):
    plt.text(y=i, x=j+0.3, s=str(j))
plt.show()
```

<b>Removing spines on right and top</b>

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8']
y = [10, 12, 15, 3, 5]
plt.barh(x, y, color=['skyblue', 'orange', 'green', 'yellow', 'pink'], alpha=0.5)
plt.xlabel('Count')
plt.ylabel('IPAddress')
plt.title('IP Address Statistics', pad=25)
plt.gca().spines['right'].set_visible(False) # remove right spine
plt.gca().spines['top'].set_visible(False) # remove left spine
for i, j in enumerate(y):
    plt.text(y=i, x=j+0.3, s=str(j))
plt.show()
```

<b>Save Image</b>

```python
x = ['1.2.3.4', '2.3.4.5', '3.4.5.6', '4.5.6.7', '5.6.7.8']
y = [10, 12, 15, 3, 5]
plt.barh(x, y, color=['skyblue', 'orange', 'green', 'yellow', 'pink'], alpha=0.5)
plt.xlabel('IPAddress')
plt.ylabel('Count')
plt.title('IP Address Statistics', pad=25)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
for i, j in enumerate(y):
    plt.text(y=i, x=j+0.3, s=str(j))
plt.savefig('your_file_path', dpi=400, bbox_inches='tight') # save the picture to specified location
```
