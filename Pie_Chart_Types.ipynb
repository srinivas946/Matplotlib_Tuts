### Basic Chart

```python
import matplotlib.pyplot as plt  # import matplotlib package and use pyplot class as plt
```

```python
ipadd = ['1.2.3.4', '2.3.4.5', '5.6.7.8', '8.9.1.0', '12.34.56.78']  # create a list of data
hits = [10, 20 , 30, 40, 50]  # create a list of ranges for respective data
plt.figure(figsize=(5, 5))    # set the size of the pie chart
# plt.axis('equal')  you can also use equal axis to set the circle shape 
plt.pie(hits, labels=ipadd)   # create pie chart with data values
plt.show()                   # show the pie chart
```

### Save Pie Chart

```python
ipadd = ['1.2.3.4', '2.3.4.5', '5.6.7.8', '8.9.1.0', '12.34.56.78']  # create a list of data
hits = [10, 20 , 30, 40, 50]  # create a list of ranges for respective data
plt.figure(figsize=(5, 5))    # set the size of the pie chart
plt.pie(hits, labels=ipadd)   # create pie chart with data values
plt.savefig('filename_to_save.png') # Save figure in png format by sepcifiying the complete path 
```

### Chart with Rotation

```python
ipadd = ['1.2.3.4', '2.3.4.5', '5.6.7.8', '8.9.1.0', '12.34.56.78']
hits = [10, 20 , 30, 40, 50]
plt.figure(figsize=(5, 5))
plt.pie(hits, labels=ipadd, startangle=90)   # startangle property used to change the orientation of pie chart
plt.show()
```

### Changing the Pie Chart Colors

```python
ipadd = ['1.2.3.4', '2.3.4.5', '5.6.7.8', '8.9.1.0', '12.34.56.78']
hits = [10, 20 , 30, 40, 50]
pie_colors = ['red', 'blue', 'green', 'orange', 'gray'] # Create a list of colors
plt.figure(figsize=(5, 5))
plt.pie(hits, labels=ipadd, colors=pie_colors, startangle=90) # assign pie_colors to colors parameter
plt.show()
```

### Changing Directions (clockwise & Counter clockwise)

```python
ipadd = ['1.2.3.4', '2.3.4.5', '5.6.7.8', '8.9.1.0', '12.34.56.78']
hits = [10, 20 , 30, 40, 50]
plt.figure(figsize=(5, 5))

# set counterclock = True to enable counter clock wise direction
# set counterclock = False to enable clock wise direction
plt.pie(hits, labels=ipadd, startangle=150, counterclock=True) 
plt.show()
```

### Add Title and Legend

```python
ipadd = ['1.2.3.4', '2.3.4.5', '5.6.7.8', '8.9.1.0', '12.34.56.78']
hits = [10, 20 , 30, 40, 50]
plt.figure(figsize=(5, 5))
plt.pie(hits, labels=ipadd, startangle=150)
plt.title('Pie Chart for IPAddress') # Add Title to the PieChart

# Add legend using legend method and use bbox_to_anchor to set the position
plt.legend(title='IPAddress', bbox_to_anchor=(1, 0, 0.5, 1), shadow=True) # if you want shadow for legend use shadow=True
plt.show()
```

### Change the size of the pie chart using radius

```python
ipadd = ['1.2.3.4', '2.3.4.5', '5.6.7.8', '8.9.1.0', '12.34.56.78']
hits = [10, 20 , 30, 40, 50]
plt.figure(figsize=(5, 5))
plt.pie(hits, labels=ipadd, startangle=150, radius=1.5) # change the radius value, default value is 1
plt.show()
```

### Percentage Label

```python
ipadd = ['1.2.3.4', '2.3.4.5', '5.6.7.8', '8.9.1.0', '12.34.56.78']
hits = [10, 20 , 30, 40, 50]
plt.figure(figsize=(5, 5))
plt.pie(hits, labels=ipadd, startangle=150, autopct='%.2f%%') # use autopct for percentage formatted string
plt.show()
```

### Percentage and Value Label

```python
ipadd = ['1.2.3.4', '2.3.4.5', '5.6.7.8', '8.9.1.0', '12.34.56.78']
hits = [10, 20 , 30, 40, 50]
plt.figure(figsize=(5, 5))

# Create a function which return the labels in the format of percentage and numbers
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)  # adding all the values
        val = int(round(pct*total/100.0))  # logic to get individual value
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)  # return the data
    return my_autopct

plt.pie(hits, labels=ipadd, startangle=150, autopct=make_autopct(hits)) # use autopct for percentage formatted string
plt.show()
```

### explode Pie Chart

```python
ipadd = ['1.2.3.4', '2.3.4.5', '5.6.7.8', '8.9.1.0', '12.34.56.78']
hits = [10, 20 , 30, 40, 50]
plt.figure(figsize=(5, 5))

# use explode to explode a particular data, explode parameter related to values (in our case hits)
plt.pie(hits, labels=ipadd, startangle=150, autopct='%.2f%%', explode=[0, 0.2, 0, 0, 0]) 
plt.show()
```

### Pie Chart Shadow

```python
ipadd = ['1.2.3.4', '2.3.4.5', '5.6.7.8', '8.9.1.0', '12.34.56.78']
hits = [10, 20 , 30, 40, 50]
plt.figure(figsize=(5, 5))
plt.pie(hits, labels=ipadd, startangle=150, autopct='%.2f%%', explode=[0, 0.2, 0, 0, 0], shadow=True) # use shadow=True 
plt.show()
```
